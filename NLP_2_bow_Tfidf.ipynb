## BOW

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
A1 = 'hello and welcome dosto' 
A2 = 'shri love NLP' 
A3 = 'shri is good boy'
```

```python
cv = CountVectorizer(stop_words='english')
vectors = cv.fit_transform([A1,A2,A3])  ## * fit() The vectorizer learns all unique words (vocabulary).
feature_names = cv.get_feature_names_out()  ## simply returns keys in index order.
dense = vectors.todense()  ## This converts a sparse matrix into a dense (normal) matrix.
```

```python
print(vectors)
```

```python
print(feature_names)
```

```python
print(dense)
```

```python
result = pd.DataFrame(dense, columns=feature_names)

print(result)
```

## TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
A1 = 'hello and welcome dosto'
A2 = 'shri love NLP'
A3 = 'shri is good boy'
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vectors = vectorizer.fit_transform([A1,A2,A3])
feature_names = vectorizer.get_feature_names_out()
dense = vectors.todense()

result = pd.DataFrame(dense, columns=feature_names)

print(result)
```

```python
print(vectorizer.idf_)
print(vectorizer.get_feature_names_out())
```
